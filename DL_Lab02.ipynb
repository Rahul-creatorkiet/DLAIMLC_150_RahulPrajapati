```python
X = [0.7, 0.6, 0.8, 0.5, 0.4, 0.9]
W = [0.2, 0.4, 0.3, 0.1, 0.2, 0.5]
b = -0.3
lr = 0.1
y = 1
```

```python
z = 0
for i in range(len(X)):
    z = z + (X[i] * W[i])

z = z + b
print(z)
```

```python
if z >= 0:
    y_pred = 1
else:
    y_pred = 0
y_pred
```

```python
error = y - y_pred
error
```

```python
if error != 0:
    for i in range(len(W)):
        W[i] = W[i] + (lr * error * X[i])
```

```python
b = b + (lr * error)
```

```python
z_new = 0
for i in range(len(X)):
    z_new = z_new + (X[i] * W[i])

z_new = z_new + b
```

```python
if z_new >= 0:
    y_new = 1
else:
    y_new = 0
```

```python
print("\nFinal Result:")
if y_new == y:
    print("Target Achieved (Pass = 1)")
else:
    print("Target Not Achieved")
```
